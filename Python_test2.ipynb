```python
a = 761
b = 159
print("a =", a, "b =", b)
print("Сумма переменных a и b:", a + b)
print ("Разность переменных a и b:", a - b)
print ("Разность переменных b и a:", b - a)
print("Произведение переменных a и b:", a * b)
print("Деление переменных a и b:", a / b)
print("Деление переменных b и a:", b / a)
print("Целочисленное деление переменных a и b:", a // b)
print("Целочисленное деление переменных b и a:", b // a)
print("a в степени b:", a ** b)
print("b в степени a:", b ** a)
```

```python
a = int(input("Введите число: "))
sum = 0
for i in range(1, a + 1):
    if i % 2 == 0:
        sum = sum + i
print("Сумма чётных чисел от 1 до", a, "равна", sum)
```
